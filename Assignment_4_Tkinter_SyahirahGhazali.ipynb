```python
# Basic Calculator to Add 2 Numbers
w = Tk()
w.title('My Simple Calculator') #the name of the widget
w.geometry('400x250') #the size

def addTwoNumbers():
    # get is a builtin method in tkinter, it's taking value from e1 and e2 (in this case)
    # entry is taking input as string by default, so need to do type casting before doing calculation
    res = int(e1.get()) + int(e2.get())
    # need to do typecasting for res too cos string concatenation need both LHS and RHS to be of the same type
    outputText = 'The Result is'+ str(res)
    outputLabel.config(text= outputText)

#Label is just a text on the widger
val1 = Label(w, text = 'Enter First Value ')
val1.place(x =30, y = 50)

val2 = Label(w, text = 'Enter Second Value ')
val2.place(x = 30, y = 90)

#Entry Button is to put in input
e1 = Entry(w)
e1.place(x=150, y=50)

e2 = Entry(w)
e2.place(x=150, y=90)

#command argument lets us to give some functionality
addButton = Button(w, text='Click Here to Add', command= addTwoNumbers)
addButton.place(x= 150, y = 120)

outputLabel = Label(w)
outputLabel.place(x= 80, y = 190)
w.mainloop()
```

```python
# Guess a number game

from tkinter import *
import random
w = Tk()
w.title('Guess A Number Game')
w.geometry('600x400')
# background colour for window
w.config(bg = '#065569')
#dont allow user to resize, set height and width to false in resizable
w.resizable(width= False, height = False)

# main Logic

ranNum = random.randint(1,10)
chance = 4
#Tkinter widgets cannot auto-update from normal Python variables
#So Tkinter uses StringVar() as a container, then connects to the widget
displayResult = StringVar()

def check_guess():
    # use global to update the value outside the function
    global ranNum
    global chance
    userInput = int(user_input.get())
    if chance > 0:
        if userInput == ranNum:
            msg = "Congratulation. You Won! "+str(ranNum)+" is the right number"
            # disable input & button so the game stop
            user_input.config(state='disabled')
            guess_button.config(state='disabled')
        elif userInput>ranNum:
            chance = chance - 1
            msg = "Your Guess Is Too High. Think Of A Smaller Number"
            user_input.delete(0,END)
        elif userInput<ranNum:
            chance = chance - 1
            msg = "Your Guess Is Too Low. Think Of A Higher Number"
            user_input.delete(0,END)
    else:
        msg = "Game Over! "+str(chance)+" Attempts Left. The Correct Number Was "+str(ranNum)
        # locks the game, limit has reached
        user_input.config(state='disabled')
        guess_button.config(state='disabled')

    displayResult.set(msg)

# Create Widgets
title = Label(w,text='Guess A Number Game',font=('Arial',20),fg='#fffcbd',bg='#065569')
gameInstruction = Label(w,text='Guess The Number Between 1 to 10 in 4 Attempts',
                        font=('Arial',13),fg='#fffcbd',bg='#065569')
user_input = Entry(w,font=('Arial',12))
guess_button = Button(w,text='Guess',font=('Arial',13),fg='#13d675',bg='black',command=check_guess)
exit_button = Button(w,text='Exit Game',font=('Arial',14),fg='white',bg='#b82741',command=w.destroy)
outputLabel = Label(w,font=('Arial',14),fg='#fffcbd',bg='#065569',textvariable=displayResult)

# Place Widgets
title.place(x=140,y=50)
gameInstruction.place(x=95,y=95)
user_input.place(x=180,y=150)
guess_button.place(x=370,y=147)
exit_button.place(x=300,y=300)
outputLabel.place(x=100,y=220)

w.mainloop()
```

```python
# Sentiment Analyzer

from tkinter import *
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sid_obj = SentimentIntensityAnalyzer()

w = Tk()
w.title('Sentiment Analyzer')
w.geometry('600x400')
#background color for window
w.config(bg = '#41644A')
w.resizable(width=False, height=False)

# main logic

displayResult = StringVar()

def check_sentiment():
    userInput = user_input.get()
    res = TextBlob(userInput)
    res = sid_obj.polarity_scores(userInput)

    if res['compound']  >= 0.5:
        displayResult.set('Positive Text')
        user_input.delete(0,END)
    elif res['compound'] <= -0.5:
        displayResult.set('Negative Text')
        user_input.delete(0,END)
    else:
        displayResult.set('Neutral Text')
        user_input.delete(0,END)

# Create Widget
title = Label(w, text = 'Sentiment Analyser Please Insert Some Text Below',
              font = ('Arial', 16), fg = '#EBE1D1' , bg ='#41644A')
user_input = Entry(w,font=('Arial', 14))
analyzer_button = Button(w, text = 'Click Here To Analyse',
                         font=('Arial', 14), fg='#EBE1D1', bg ='#E9762B', command=check_sentiment)
output_label = Label(w, font = ('Arial', 16), fg = '#F1E6C9' , bg ='#41644A',
                    textvariable=displayResult)

# Place widget
title.place(x=50, y=30)
user_input.place(x=180, y=80)
analyzer_button.place(x=190, y=120)
output_label.place(x=190, y=200)

w.mainloop()
```

```python
from tkinter import *
import wolframalpha
client = wolframalpha.Client('LUHHQYJY76') #an object of a class using the unique app id from wolframalpha

w = Tk()
w.title('Wolfram Alpha')
w.geometry('800x400')
#background color for window
w.config(bg = '#41644A')
#w.resizable(width=False, height=False)

displayAnswer = StringVar()

def check_answer():

  # Send the text to WolframAlpha and wait for a response
    query = user_input.get().strip()

    if not query:
        displayAnswer.set("Please enter a question")
        return

    if query.lower() in ['exit', 'quit', 'stop']:
        w.destroy()
        return

    try:
        res = client.query(query)
        answer = next(res.results).text  # Get first usable answer
        displayAnswer.set(answer)
    except StopIteration:
        displayAnswer.set("No answer found")
    except Exception as e:
        displayAnswer.set(f"Error: {e}")

def exit():
    w.destroy()

# Create Widget
title = Label(w, text = 'Ask A Question',
              font = ('Arial', 16), fg = '#EBE1D1' , bg ='#41644A')
user_input = Entry(w,font=('Arial', 14))
answer_button = Button(w, text = 'Answer',
                         font=('Arial', 14), fg='#EBE1D1', bg ='#E9762B', command=check_answer)
output_label = Label(w, font = ('Arial', 16), fg = '#F1E6C9' , bg ='#41644A',
                    textvariable=displayAnswer)
exit_button = Button(w, text = 'Exit',
                         font=('Arial', 14), fg='#EBE1D1', bg ='#E9762B', command=exit)

# Place widget
title.place(x=50, y=30)
user_input.place(x=180, y=80)
answer_button.place(x=190, y=120)
output_label.place(x=190, y=200)
exit_button.place(x=320, y=120)

w.mainloop()
```
