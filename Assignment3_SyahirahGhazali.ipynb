```python
# Q1. Python program to count positive and negative numbers in a list.
# a = [10, -20, 30, -40, 50, -60, 0]

a = [10, -20, 30, -40, 50, -60, 0]
# no of negative numbers
neg = 0
# no of positive numbers
pos = 0

for i in a:
  if i<0:
    neg+=1
  elif i>0:
    pos+=1

print(neg)
print(pos)
```

```python
# Q2. Using List Comprehension. Write a Python program to count positive and negative numbers in a list.
# a = [10, -20, 30, -40, 50, -60, 0]

neg = [i for i in a if i<0]
pos = [i for i in a if i>0]

print(len(neg))
print(len(pos))
```

```python
# Q3. Remove Multiple Elements from List in Python
# a = [10, 20, 30, 40, 50, 60, 70]
# remove = [20, 40, 60]
# Resulting list = [10, 30, 50, 70]

a = [10, 20, 30, 40, 50, 60, 70]
result = list(filter(lambda x: x not in [20, 40, 60], a))
print(result)
```

```python
# Q6. Key with Maximum Unique Values - Python
# dict1 = {"key1": [5, 7, 5, 4, 5], "key2": [6, 7, 4, 3, 3], "key3": [9, 9, 6, 5, 5]}
# output : key2

dict1 = {"key 1": [5, 7, 5, 4, 5], "key 2": [6, 7, 4, 3, 3], "key 3": [9, 9, 6, 5, 5]}

max_key, max_val = None, 0

for k, v in dict1.items():
  unique_count = 0                      # reset unique counter to 0 for each key
  for i in v:                           # 5
    if v.count(i) == 1:                 # using count method to find how many times 5 occured, if 5 occured only once, then count in unique values
      unique_count+=1
    print(f'{k}: {unique_count}')       # just checking the counter if it works
    if unique_count > max_val:          # if unique count for key 2 > key 1
       max_val = unique_count           # update max unique val to key 2 unique value
       max_key = k                      # update max key to key 2, else remain

print(f'the key that has maximum value is {max_key} with {max_val} unique values')
```

```python
# Q7. Story: An online store records daily product sales.
# Convert this list into total sales per product.
sales = [
    ("apple", 3),
    ("banana", 2),
    ("apple", 5),
    ("banana", 1),
    ("orange", 4)
]

total_sales = {}
for product, quantity in sales:
    if product in total_sales:
        total_sales[product] += quantity     # count if product is in total_sales
    else:
        total_sales[product] = quantity      # add product to total_sales


print(total_sales)

# Output : {'apple': 8, 'banana': 3, 'orange': 4}
```

```python
# Q8. Undo Feature in Text Editor
# A text editor supports an undo operation.
# Given operations:
# actions = ["type A", "type B", "delete", "type C"]
# Output : ['type A', 'type C']

# actions = ["type A", "type B", "delete", "type C","type B","delete","type A","type C","delete"]
# Output : ['type A', 'type C', 'type A']

actions = ["type A", "type B", "delete", "type C"]
actions2 = ["type A", "type B", "delete", "type C","type B","delete","type A","type C","delete"]

def undo(actions):
  undo_stack = []
  for action in actions:
      if action == "delete":
          if undo_stack != []:                # if undo_stack is not empty. alternatively, 1) if len(undo_stack)>0 or, 2) if undo_stack:
              undo_stack.pop()
      else:                                   # if it's not delete, then append to the stack
          undo_stack.append(action)
  return undo_stack

undo_stack = undo(actions)
undo_stack2 = undo(actions2)

print(undo_stack)
print(undo_stack2)
```

```python
# Q9. Print the index number of those Two boxes whose weights sum to a target.
# weights = [2, 7, 11, 15 , 6 , 3, 7 , 3 , 7 , 2]
# target = 9

# Output
'''
0 1
4 5
0 6
4 7
0 8
8 9
'''

weights = [2, 7, 11, 15 , 6 , 3, 7 , 3 , 7 , 2]
target = 9
seen = {}

for index, weight in enumerate(weights):             # index = 0, weight = 2                    # index = 1, weight = 7
    if target - weight in seen:                      # 9-2 is not in seen dictionary            # 9-7 is in seen key
        print(seen[target - weight], index)                                                     # print seen[2] which is 0 and index which is 1
    seen[weight] = index                             # seen[2] = 0                              # seen[7] = 1
    #print(seen)                                     # seen = {2: 0}                            # seen = {2: 0, 7: 1}
```

```python
# Q10. Online Store Price Update (map + lambda)
# An online store increases all product prices by 10%.
# prices = [100, 200, 300, 400]
# output [110, 220, 330, 440]

prices = [100, 200, 300, 400]
new_prices = list(map(lambda x: x*0.1+x, prices))
print(new_prices)
```

```python
# Q11. Employee Bonus System (map + if–else)
# Employees earning ≥ 50,000 get a 20% bonus, others get 10%.
# salaries = [40000, 60000, 50000, 30000]

salaries = [40000, 60000, 50000, 30000]
new_salaries = list(map(lambda x: x*1.2 if x>=50000 else x*1.1, salaries))
print(new_salaries)
```

```python
# Q12. Order Filtering System (filter + multiple conditions)
# An e-commerce system wants orders above ₹1000 and not cancelled.

orders = [
    {"amount": 1200, "status": "completed"},
    {"amount": 800, "status": "completed"},
    {"amount": 1500, "status": "cancelled"},
    {"amount": 2000, "status": "completed"}
    ]

filtered_orders = list(filter(lambda x: x["amount"] > 1000 and x["status"] == "completed", orders))
print(filtered_orders)
```

```python
# Q13. Data Cleaning Pipeline (map + filter + lambda)
# From a data pipeline:
#  a.)Remove negative numbers
#  b.) Square remaining numbers
#  c.) Keep only results > 50

# data = [-4, -1, 2, 4, 8]
# Output : [64]

data = [-4, -1, 2, 4, 8]
data = list(filter(lambda x: x>0, data))
print(data)
data = list(map(lambda x: x**2, data))
print(data)
data = list(filter(lambda x: x>50, data))
print(data)
```
